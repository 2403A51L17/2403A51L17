```python
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Task 1: Load and Visualize the Data
# Generate synthetic dataset
np.random.seed(0)  # Different seed for variety
X = np.linspace(-5, 5, 100).reshape(-1, 1)  # 100 points from -5 to 5
y = 1 + 2 * X + 0.3 * X**2 + np.random.normal(0, 1.5, X.shape)  # Quadratic with noise

# Initial visualization
plt.figure(figsize=(8, 6))
plt.scatter(X, y, color='green', label='Data Points', alpha=0.6)
plt.title('Initial Data Exploration')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()

# Task 2: Apply Polynomial Regression
# Transform features to polynomial (degree 2)
poly_transformer = PolynomialFeatures(degree=2, include_bias=False)  # Exclude bias for clarity
X_transformed = poly_transformer.fit_transform(X)  # [X, X^2]

# Train the model
poly_model = LinearRegression()
poly_model.fit(X_transformed, y)

# Task 3: Evaluate and Visualize
# Generate predictions
y_fitted = poly_model.predict(X_transformed)

# Calculate performance metrics
error = mean_squared_error(y, y_fitted)
r2_value = r2_score(y, y_fitted)
print(f"Model Performance:")
print(f"Mean Squared Error: {error:.3f}")
print(f"R² Score: {r2_value:.3f}")

# Visualization of fit
plt.figure(figsize=(8, 6))
plt.scatter(X, y, color='green', label='Data Points', alpha=0.6)
plt.plot(X, y_fitted, color='purple', label='Polynomial Fit (Degree 2)', linewidth=2)
plt.title('Polynomial Regression Result')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()

# Display model details
print("\nModel Details:")
print(f"Coefficients: {poly_model.coef_}")
print(f"Intercept: {poly_model.intercept_[0]:.3f}") # Access the first element of the intercept array
```
